```python
#Calculate the mean and median of Page Views and Time Spent for both groups (A and B)
df["Group"].unique()
```

```python
grouped=df.groupby("Group")
```

```python
data=grouped[["Page Views","Time Spent"]].agg(["mean","median"])
data
```

```python
#Which group has a higher conversion rate? Is the difference significant?
```

```python
total_yes=df.groupby("Group")["Conversion"].apply(lambda i: (i=="Yes").sum() )
```

```python
total_yes
```

```python
total_user=df.count()
total_user
```

```python
users = df['Group'].value_counts()
users
```

```python
conversion_rate=(total_yes/users)
conversion_rate=conversion_rate*100
conversion_rate
```

```python
group_A=conversion_rate["A"]
group_B=conversion_rate["B"]
```

```python
output = f"the Group A conversion rate is {group_A:.2f}% and the Group B conversion rate is {group_B:.2f}% "
print(output)
```

```python
#Find the total number of users and the conversion rate for each location.
```

```python
df
```

```python
grouped=df.groupby("Location")
grouped
```

```python
convertion=grouped["Conversion"].apply(lambda i : (i=="Yes").sum())
convertion
```

```python
users_count=grouped["User ID"].count()
users_count
```

```python
conversion_rate=(convertion/users_count)*100
conversion_rate
```

```python
result = pd.DataFrame({
    "Total Users": users_count,
    "Total Conversions": convertion,
    "Conversion Rate (%)": conversion_rate
})
```

```python
print(result)
```
